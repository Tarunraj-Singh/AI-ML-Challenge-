# Importing the necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("diabetes_012_health_indicators_BRFSS2021.csv")
df.head(10)
```

# Checking and Preprocessing the Dataset

```python
print(df.info())
```

```python
#As we can see that there is no null value in the dataset. Still for our convinience, we can do:-
df.dropna(inplace = True)     # As we have ample data, we can directly drop the data instead of replacing it with zero
```

```python
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
#Removing Duplicates
df.drop_duplicates(inplace = True)
df.info()
```

```python
#As every data in the dataset contains 0 after decimal point. So we can consider to convert it into int type 
#for low memory consumption
#In real world,BMI is in float format, So we should not convert it to int
for column in df.columns:
    if column != 'BMI':
        df[column] = df[column].astype(int)
df.info()
```

```python
#As we are only interested in knowing that whether the person has diabetes or not rather than the type of diabetes
#We can change the value of from 2 to 1 and make it binary
df['Diabetes_012'] = df['Diabetes_012'].replace(2, 1)
```

```python
for column in df.columns:
    unique_values_sum = df[column].nunique()
    print(column  , ":", unique_values_sum)
```

# Building the model

```python
y = df["Diabetes_012"]
X = df.drop(columns=['Diabetes_012'])
```

```python
#scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(random_state=42)
clf.fit(X_train, y_train)
```

# Evaluation and Tunning

```python
# Predictions on training set
train_pred = clf.predict(X_train)

# Predictions on testing set
test_pred = clf.predict(X_test)

# Calculate metrics on training set
train_accuracy = accuracy_score(y_train, train_pred)
train_precision = precision_score(y_train, train_pred)
train_recall = recall_score(y_train, train_pred)
train_f1 = f1_score(y_train, train_pred)

# Calculate metrics on testing set
test_accuracy = accuracy_score(y_test, test_pred)
test_precision = precision_score(y_test, test_pred)
test_recall = recall_score(y_test, test_pred)
test_f1 = f1_score(y_test, test_pred)

# Print metrics
print("Training Report:")
print(f"Accuracy: {train_accuracy:.2f}")
print(f"Precision: {train_precision:.2f}")
print(f"Recall: {train_recall:.2f}")
print(f"F1-score: {train_f1:.2f}")
print("\nTesting Reprt:")
print(f"Accuracy: {test_accuracy:.2f}")
print(f"Precision: {test_precision:.2f}")
print(f"Recall: {test_recall:.2f}")
print(f"F1-score: {test_f1:.2f}")
```

```python
# parameter grid
param_grid = {
    'penalty': ['l1', 'l2'],              
    'C': [0.001, 0.01, 0.1, 1, 10, 100]   
}

# Logistic Regression
logistic_regression = LogisticRegression(random_state=42, max_iter=1000)

# GridSearchCV
grid_search = GridSearchCV(logistic_regression, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Perform GridSearchCV
grid_search.fit(X_train, y_train)

# Best Parameters and Best estimator
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

best_test_accuracy = best_model.score(X_test, y_test)

print("Best parameters:", best_params)
print("Best test accuracy:", best_test_accuracy)
     
```

# Rebuilding the final model

```python
# Initialize logistic regression with best parameters
final_model = LogisticRegression(C=0.1, penalty='l2')

# Train the final model on the entire training dataset
final_model.fit(X_train, y_train)
```

```python
# Calculate metrics on training set
train_pred = final_model.predict(X_train)
train_accuracy = accuracy_score(y_train, train_pred)
train_precision = precision_score(y_train, train_pred)
train_recall = recall_score(y_train, train_pred)
train_f1 = f1_score(y_train, train_pred)

# Calculate metrics on testing set
test_pred = final_model.predict(X_test)
test_accuracy = accuracy_score(y_test, test_pred)
test_precision = precision_score(y_test, test_pred)
test_recall = recall_score(y_test, test_pred)
test_f1 = f1_score(y_test, test_pred)

# Print metrics
print("Training Report:")
print(f"Accuracy: {train_accuracy:.2f}")
print(f"Precision: {train_precision:.2f}")
print(f"Recall: {train_recall:.2f}")
print(f"F1-score: {train_f1:.2f}")
print("\nTesting Report:")
print(f"Accuracy: {test_accuracy:.2f}")
print(f"Precision: {test_precision:.2f}")
print(f"Recall: {test_recall:.2f}")
print(f"F1-score: {test_f1:.2f}")

```
